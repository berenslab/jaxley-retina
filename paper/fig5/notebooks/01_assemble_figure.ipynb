```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

```python
with open("fig5_0504_209.pkl", "rb") as f:
    data = pickle.load(f)
gs_by_readout = data["conductances"]
coords = data["coords"]
```

```python
with mpl.rc_context(fname="../../.matplotlibrc"):
    fig, axs = plt.subplots(2, 5, figsize=(5.5, 2.25), layout="constrained")
    for i, ax in enumerate(axs.ravel()):
        im = ax.scatter(coords[0], coords[1], c=np.log(gs_by_readout[i]), cmap="binary", vmin=-20, vmax=-3)
        ax.set_title(f"Readout {i}", pad=0)
        ax.axis("off")
        ax.axis("off")
        ax.set_aspect("equal")
        ax.set_xlim((-2900, 2900))
        ax.set_ylim((-2900, 2900))
    cbar = fig.colorbar(im, ax=axs.ravel().tolist(), aspect=30)
    cbar.set_label(r"$\log(g_s) \, \, (\mu S)$", labelpad=-10)
    cbar.set_ticks([-3, -20])

    fig.text(0, 1.05, "Ribbon synapse conductances onto each of the readout neurons", ha="left", fontsize=7)
    
    plt.savefig(f"../fig/fig5.png", dpi=300)
    plt.savefig("../fig/fig5.pdf", format="pdf", bbox_inches="tight")
    plt.show()

```
